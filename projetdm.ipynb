```python
demo_features = ['HeartDisease', 'Smoking', 'Sex', 'AlcoholDrinking']
```

```python
plt.figure(figsize=(15,12))

for i,col in enumerate(demo_features):    
    plt.subplot(2,2,i + 1)
    
    ax = sns.countplot(data = df, x = col, palette = 'rocket')

    plt.title('Distribution des personnes par ' + col)
    plt.xlabel(col, fontsize= 14)
    plt.ylabel("Nombre de personnes", fontsize= 14)
    plt.ylim(0,200000)
    plt.xticks(fontsize= 15)
    plt.yticks(fontsize= 14)

    for p in ax.patches:
        ax.annotate((p.get_height()), (p.get_x()+0.32, p.get_height()+300), fontsize= 13)

plt.tight_layout()

plt.show()
```

```python
num_accinfo_features=['MentalHealth','PhysicalHealth','BMI']
```

```python
plt.figure(figsize=(12,15))

for i, col in enumerate(num_accinfo_features):
  plt.subplot(3,1,i + 1)
  plt.ylim(0,200000)
  ax = sns.histplot(data=df, x=col)
  ax.set_xlabel(col)
  ax.set_ylabel('Nombre de clients')
  ax.set_title(f'Distribution des personnes selon la variable '+col)

  for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.32, p.get_height()+50), fontsize= 13)

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(12,15))

for i,col in enumerate(num_accinfo_features):    
    plt.subplot(3,1,i + 1)
    sns.distplot(df.loc[:,col])
    plt.title('Distribution des personnes selon '+col)
    plt.ylabel('AgeCategory')
    plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(12,15))

for i, col in enumerate(num_accinfo_features):
  plt.subplot(3,1,i + 1)
  ax = sns.boxplot(data=df, x=col, color='#BBD2E1')
  ax.set_xlabel(col)
  ax.set_ylabel(None)
  ax.set_title('Les valeurs de la variable '+col)


plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(12,15))

for i,col in enumerate(num_accinfo_features):    
    plt.subplot(3,1,i + 1)    
    sns.kdeplot(df.loc[(df['AgeCategory'] == 0), col], label = 'No ', shade = True)
    sns.kdeplot(df.loc[(df['AgeCategory'] == 1), col], label = 'Yes', shade = True)
    plt.title('Concentration du churn selon '+ col)
    plt.legend()
    plt.ylabel('')
    plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(12,15))

ax = sns.lmplot(x='AgeCategory',y='Sex', data=df, height=6, aspect=2)

plt.title(f'Correlation entre les variables Tenure in Months et Total Charge')

plt.tight_layout()
plt.show()
```
